```python
#Probability of Default model - PD estimation
import numpy as np
import scipy.stats as stats 
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import random
from random import seed
from random import randint
from scipy.stats import norm
from datetime import date
from scipy.stats import norm
import matplotlib.ticker as mtick

import seaborn as sns

import statsmodels.api as sm

from sklearn.preprocessing import StandardScaler
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
from sklearn import metrics as mt
from sklearn import svm


from math import log,sqrt
import seaborn as sns
sns.set_style("white")

%matplotlib inline
#Load data
file = 'corporate_credit_risk.csv'
db_cr = pd.read_csv(file)
```

```python
#Function for the Accuracy Ratio: trata de medir qué tanto se ajusta tu modelo de PD (modelo discriminatorio) a la default rate observada
#metrica va de 0-1, se espera q un buen modelo tenga AR del 60% esta bien, si da 90% puede que tengas algo mal
def AR(pd,default,alpha):
    D = np.array(pd[default==1])
    ND = np.array(pd[default==0])
    NDe = len(D) #numbers of defaults
    NND = len(ND) #number of none defaults
    UDND = np.array([])
    
    #Calculus for the U-estimator
    for i in range(0,NDe):
        aux = np.where(D[i]>ND,1,
                       np.where(D[i]==ND,.5,0))
        
        UDND = np.append(UDND,np.array(aux))
        
    #U-estimator
    Uhat = sum(UDND/(NND*NDe))
        
    #parameters for the variance
    Pxxy = Uhat/(2-Uhat)
    Pxyy = 2*Uhat**2/(1+Uhat)
        
    #Var for the AUC
    var = (Uhat*(1-Uhat)+(NDe-1)*(Pxxy-Uhat**2)+(NND-1)*(Pxyy-Uhat**2))/(NDe*NND)
        
    #Confidence interval for the AR
    AR = 2*Uhat-1
    var_ar = 4*var
    CIL = AR -np.sqrt(var_ar)*norm.ppf(1-alpha/2)
    CIR = AR +np.sqrt(var_ar)*norm.ppf(1-alpha/2)
    values = [AR,var_ar,CIL,CIR]
    
    return values
```

```python
def corr_function(db):
    corr_df = db.corr().round(2)
    max_corr = 0.4
    mask = np.zeros_like(corr_df)
    mask[np.triu_indices_from(mask)] =  True

    with sns.axes_style("white"):
        f,ax = plt.subplots(figsize=(15,15))
        ax = sns.heatmap(corr_df,mask=mask,vmax=max_corr,square=True,annot=True,cmap='YlGnBu')
```

```python
#Analysis of correlations
corr_function(X_num) #gráfica para ver qué tanto están correlacionadas mis variables 
#mas negro significa que estan más correlacionadas 
```

```python
#Model estimation with logistic regression
#Utilizamos Machine Learning
#utilizamos ref¿gresión logística porque se que mi varibale es dicotómica 
#da regresión L. individaules para cada una de las variables, hace simulaciones con distintos ampleos (400) de la muestra
#si le dices que se quede con los mejores x 
rfe_selector = RFE(estimator=LogisticRegression(solver='lbfgs', max_iter=400), step = 1)
rfe_selector.fit(X_train, y_train)
predicteve_columns = list(X_train.columns[rfe_selector.get_support()])
predicteve_columns
#pasar a los mejores 15 drivers
#se queda abirto para que el algoritmo me de las mejores variables dice: "Del conjunto de variables que me acabas de pasas, haciendo modelos de RL estos son los mejores drivers"
```

```python
X_train_v2 = X_train[predicteve_columns] 
#base train, base con la que voy a desarrollar mi modelo, las evaluo con las columnas que me dijo el feature selection
X_train_v2 = sm.add_constant(X_train_v2,prepend=True)
#le agrego la columna 2 que será la beta 0
modelo = sm.Logit(endog=y_train,exog=X_train_v2).fit(disp=False)
print(modelo.summary())
#mi análisis ANOBA es el siguiente:
#ya que corrí el modelo, teca ver que variable esta siendo más significativa, drivers que tengan p-vaule >5% son estadírticamnete no significativos
#es decir, los p-value mayores a 5% los saco
#toca excluir esos drivers
```

```python
def model_summary(modelo):
    df = modelo.params.reset_index()
    df.columns=['Value','Coeff']
    df['Sign'] = np.where(df['Coeff']<0,'Negative','Positive')
    df['P-values']=[round(x,4) for x in list(modelo.pvalues)]
    return df
```

```python
#analizmaos el signo del negocio
summary_model = model_summary(modelo)
summary_model
#la constante siempre va adentro, asi que esa no se contempla 
#entre más gastos tiene una empresa, mayor probabilidad de incumplimiento 
#correlacion positiva; si cambia una variable, también cambia la otra 
```

```python
def performance_metrics(y_hat,y):
    AccR = AR(y_hat,y,.05)
    fpr,tpr,_ = mt.roc_curve(y,y_hat)
    AUC = mt.auc(fpr,tpr)
    metrics = pd.DataFrame({'AR':[AccR[0]],'AUC':[AUC]})
    return metrics
```

```python
y_hat_1 = modelo.predict(X_train_v5)
y_hat_2 = modelo2.predict(X_train_v5.drop(['_ Not Audited'],axis=1))
#un modelo tiene la variable de auditada y la otra no lo tiene
```

```python
performance_metrics(y_hat_1,y_train) #1° modelo
```

```python
performance_metrics(y_hat_2,y_train) #2° modelo
```

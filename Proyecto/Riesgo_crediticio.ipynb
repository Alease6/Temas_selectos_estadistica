```python
#Function for the Accuracy Ratio: trata de medir quÃ© tanto se ajusta tu modelo de PD (modelo discriminatorio) a la default rate observada
#metrica va de 0-1, se espera q un buen modelo tenga AR del 60% esta bien, si da 90% puede que tengas algo mal
def AR(pd,default,alpha):
    D = np.array(pd[default==1])
    ND = np.array(pd[default==0])
    NDe = len(D) #numbers of defaults
    NND = len(ND) #number of none defaults
    UDND = np.array([])
    
    #Calculus for the U-estimator
    for i in range(0,NDe):
        aux = np.where(D[i]>ND,1,
                       np.where(D[i]==ND,.5,0))
        
        UDND = np.append(UDND,np.array(aux))
        
    #U-estimator
    Uhat = sum(UDND/(NND*NDe))
        
    #parameters for the variance
    Pxxy = Uhat/(2-Uhat)
    Pxyy = 2*Uhat**2/(1+Uhat)
        
    #Var for the AUC
    var = (Uhat*(1-Uhat)+(NDe-1)*(Pxxy-Uhat**2)+(NND-1)*(Pxyy-Uhat**2))/(NDe*NND)
        
    #Confidence interval for the AR
    AR = 2*Uhat-1
    var_ar = 4*var
    CIL = AR -np.sqrt(var_ar)*norm.ppf(1-alpha/2)
    CIR = AR +np.sqrt(var_ar)*norm.ppf(1-alpha/2)
    values = [AR,var_ar,CIL,CIR]
    
    return values
```

```python
def corr_function(db):
    corr_df = db.corr().round(2)
    max_corr = 0.4
    mask = np.zeros_like(corr_df)
    mask[np.triu_indices_from(mask)] =  True

    with sns.axes_style("white"):
        f,ax = plt.subplots(figsize=(15,15))
        ax = sns.heatmap(corr_df,mask=mask,vmax=max_corr,square=True,annot=True,cmap='YlGnBu')
```

```python
def model_summary(modelo):
    df = modelo.params.reset_index()
    df.columns=['Value','Coeff']
    df['Sign'] = np.where(df['Coeff']<0,'Negative','Positive')
    df['P-values']=[round(x,4) for x in list(modelo.pvalues)]
    return df
```

```python
def performance_metrics(y_hat,y):
    AccR = AR(y_hat,y,.05)
    fpr,tpr,_ = mt.roc_curve(y,y_hat)
    AUC = mt.auc(fpr,tpr)
    metrics = pd.DataFrame({'AR':[AccR[0]],'AUC':[AUC]})
    return metrics
```
